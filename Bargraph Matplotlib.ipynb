```python
import numpy as np
import matplotlib.pyplot as mp
```

```python
#bar graph
%matplotlib notebook
topfilms=['f1','f2','f3','f4','f5']
awards=[5,2,9,1,6]
mp.bar(range(len(topfilms)),awards)
mp.ylabel('No. Of Awards')
mp.xlabel('Top films')
###to label each bar
mp.xticks(range(len(topfilms)),topfilms)
#mp.yticks(range(len(awards)),awards)
mp.show()
```
